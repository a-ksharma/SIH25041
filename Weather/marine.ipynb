```python
# import requests
import json

url = "https://marine-weather-api-apiverve.p.rapidapi.com/v1/marineweather"

headers = {
    # Replace with your actual API key
	"x-rapidapi-key": "9e707b4dccmsha526964ad0f44e0p10fab6jsn3b4c3979c03e",
	"x-rapidapi-host": "marine-weather-api-apiverve.p.rapidapi.com",
    "Accept": "application/json"
}

# A list of coordinates for key locations in the Indian Ocean


# Dictionary to store all our results
ocean_weather_data = {}

for location_name, coords in locations.items():
    print(f"Fetching weather for {location_name}...")
    try:
        response = requests.get(url, headers=headers, params=coords, timeout=10)
        response.raise_for_status()  # Raises an HTTPError for bad responses (4xx or 5xx)
        ocean_weather_data[location_name] = response.json()
    except requests.exceptions.RequestException as e:
        print(f"Could not retrieve data for {location_name}: {e}")

# Print the collected data in a readable format
print("\n--- Collected Weather Data ---")
with open('indian_ocean_weather_data.json', 'w') as f:
    json.dump(ocean_weather_data, f, indent=2)
```
